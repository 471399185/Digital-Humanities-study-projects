### Task 3
Compute and print the readability index of the text. Use the formula of *Text Comprehension Coefficient* (*Teksta uztveramības koeficients*, TUK):

> TUK = round(A / B + C / A * 100),

where A is the number of words in the text, B is the number of sentences in the text, C is the number of words longer than 6 characters.

```python
# the number of words longer than 6 characters
six_cha_words = 0
for word in range(0, len(words)):
    for six in range(0,len(words[word])):
        if six > 6:
            six_cha_words = six_cha_words +1
print("Number of words longer than six:", six_cha_words)

#Text Comprehension Coefficient
TUK = round((len(words)/sentence_counts) + (six_cha_words/len(words)) *100)
print("Value of TUK:", TUK)
```

### Task 7
One of the ways how to compute the Lexical Richness (LR) of a text is dividing the number of unique words by the total number of words. Unfortunately, this is not really a completely correct way how to do it because LR is also influenced by the length of the text. A more correct way for computing LR is using a so-called <a href="https://en.wikipedia.org/wiki/Moving_average" target="_blank">*moving average*</a>, i.e., moving a fixed-width (say, 100 words) window over the text with a step of one word and computing LR in each such window and then finally computing mean of LR values from all such windows. Here is how to do it step-by-step:<br>
1. Define window width `n = 100`.
2. Initialize a variable for summing LR values from all iterations of the loop (described in Step 3.2).
3. Use `for` loop to loop through word indices `i` from `0` to `len(stems) - n`.<br>
   (Note that we will use __stems__ instead of words so that the LR value is less influenced by the inflectional morphology of the language which would make the lexicon seem richer than it actually is.)<br>
   In each iteration of the loop do Steps 3.1 and 3.2:<br>
   3.1. Get a 'window' from the list of stems for computing LR. The 'window' is simply a subset of your list of stems which starts at index `i` (inclusive) and ends with index `i + n` (exclusive).<br>
   3.2. Compute LR by dividing the number of unique words in the window by the window size `n` and add it to your variable initialized in Step 2.<br>
4. After the loop, compute mean of the LR to get Mean Lexical Richness (MLR). Now print it out.

```python
n = 100
lr = 0
numbers_of_lr = 0
window = set()
for i in range(0, len(stems)-n, n):
    for y in range(n):
        window.add(stems[y])
    lr +=(len(window)/n)
    numbers_of_lr +=1
print("Mean of LR:", lr/numbers_of_lr)
```

By the way:
> Vocabulary richness measurement is one of the oldest and most traditional fields in quantitative linguistics. The concept of vocabulary richness measure is based on the fact that each person uses a specific individual vocabulary. Linguists use the concept of vocabulary richness mostly in authorship and genre analysis. <a href="https://www.researchgate.net/publication/258518594_Vocabulary_Richness_Measure_in_Genres" target="_blank">(Kubát &amp; Milička, 2013)</a>
